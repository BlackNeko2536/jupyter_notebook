```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
```

```python
X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sn
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
train_X, test_X, train_y, test_y = train_test_split(X_pca,y,stratify=y)
```

```python
train_y.value_counts()
```

```python
test_y.value_counts()
```

```python
for i in range(40):
    knn = KNeighborsClassifier(n_neighbors=1+i)
    knn.fit(train_X,train_y)
    preds=knn.predict(test_X)
    print(i)
    print(confusion_matrix(test_y, preds))
    #print(i+1,knn.score(test_X,test_y))
```

```python
knn = KNeighborsClassifier(n_neighbors=13)
```

```python
knn.fit(train_X,train_y)
```

```python
print(knn.score(test_X,test_y))
```

```python
preds=knn.predict(test_X)
```

```python
confusion_matrix(test_y, preds)
```

```python
confusion_matrix = pd.crosstab(test_y,preds, rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(confusion_matrix, annot=True)
```

```python
print('Accuracy: ',metrics.accuracy_score(test_y,preds))
plt.show()
```
